# Data Structures in Python


## Array

image.png

**Definition**
An array is a collection of elements, each identified by an index or a key as we say it. In Python, arrays are provided by the 'array' module which defines a sequence data type called 'array' only.
It is a linear data structure used to store multiple item of same datatype in continuous memory location.

**Characterstics**
Arrays only contain elements of one single data type. This is the key distinction from other data structure such as lists in python, which can store elements of different data types. Another characterstic involves 'fixed size'. Arrays in python have a fixed size upon creation, which cannot be changes later on. This is also a limitation of this data structure.

**Advantages**
There are certain advantages of this data structure including 'Efficient Memory Usage' and 'Faster Access of elements through indexes'.

**Limitations**
However, there are also some limitations of this data structure which includes 'Fixed size of an array' as we cannot change the size of an array dynamically.
In addition to this, the fact that we can only store data of only one data type in one array is also a limitation of this data structure as it limits us in using these in different applications.
